```python
from skimage.io import imread
import matplotlib.pyplot as plt
params = {
    'x_min' : 200,
    'x_max' : 600,
    'y_min' : 450,
    'y_max' : 750,
    'p' : 13
    }
orig_img = imread('sMEUwaY11o0.jpg')

plt.imshow(orig_img)
plt.show()
```

```python
def blur_shreks_eyes(pic, x_min, x_max, y_min, y_max, p):

    blurred_pic = pic.copy()

    ranges_x = range(x_min, x_max+1, (x_max - x_min)//p)
    ranges_y = range(y_max, y_min - 1, -1 *(y_max - y_min)//p)

    for i in range(len(ranges_y)-1):
        for j in range(len(ranges_x)-1):
            for k in range(0,3):
                blurred_pic[ranges_y[i+1]:ranges_y[i],
                            ranges_x[j]:ranges_x[j+1],
                            k]=orig_img[ranges_y[i+1]:ranges_y[i],
                                        ranges_x[j]:ranges_x[j+1],
                                        k].mean()
    return blurred_pic
```

```python
blurred_pic = blur_shreks_eyes(orig_img, **params)
plt.imshow(blurred_pic)
plt.show()
```
